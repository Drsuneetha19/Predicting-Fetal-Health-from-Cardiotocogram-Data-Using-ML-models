**Capping & Flooring of Outliers:**

```python
#import pandas as pd
# Calculating the IQR for each column in the DataFrame
Q1 = df.quantile(0.10)
Q3 = df.quantile(0.90)
IQR = Q3 - Q1
# Defining the upper and lower bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Capping and Flooring the outliers
df_capped = df.apply(lambda x: x.where(x >= lower_bound[x.name], lower_bound[x.name]), axis=0)
df_capped = df_capped.apply(lambda x: x.where(x <= upper_bound[x.name], upper_bound[x.name]), axis=0)
# Checking changes
print(df_capped.describe())
```

**Outliers before and After Capping & Flooring method:**

```python

def count_outliers(series, lower_bound, upper_bound):
    return ((series < lower_bound) | (series > upper_bound)).sum()

# Counting outliers in the original DataFrame
outliers_original = df.apply(lambda x: count_outliers(x, lower_bound[x.name], upper_bound[x.name]))

# Counting outliers in the capped DataFrame
outliers_capped = df_capped.apply(lambda x: count_outliers(x, lower_bound[x.name], upper_bound[x.name]))

print("Outliers in original DataFrame:")
print(outliers_original)

print("\nOutliers in capped DataFrame:")
print(outliers_capped)
```

**SHAPIRO WILK TEST FOR NORMALITY:**

```python
from scipy.stats import shapiro

# List of continuous variables
fetalh = ['baseline value', 'accelerations', 'fetal_movement', 'uterine_contractions',
                   'light_decelerations', 'severe_decelerations', 'prolongued_decelerations',
                   'abnormal_short_term_variability', 'mean_value_of_short_term_variability',
                   'percentage_of_time_with_abnormal_long_term_variability', 'mean_value_of_long_term_variability','histogram_width',
       'histogram_min', 'histogram_max', 'histogram_number_of_peaks','histogram_number_of_zeroes', 'histogram_mode', 'histogram_mean',
       'histogram_median', 'histogram_variance', 'histogram_tendency','fetal_health']

# Perform Shapiro-Wilk test for each variable
normality_results = {}
for var in fetalh:
    stat, p_value = shapiro(df[var])
    normality_results[var] = {'Statistic': stat, 'p-value': p_value, 'Normality': 'Normal' if p_value > 0.05 else 'Not normal'}

# Print the results
for var, result in normality_results.items():
    print(f"Shapiro-Wilk test result for '{var}':")
    print(f"  - Test statistic: {result['Statistic']:.4f}")
    print(f"  - p-value: {result['p-value']:.4f}")
    print(f"  - Normality: {result['Normality']}\n")
```

 The Shapiro-Wilk test results indicate that all variables in the dataset significantly deviate from a normal distribution, as evidenced by extremely low p-values. Consequently, parametric methods that assume normality are unsuitable for this data. Non-parametric approaches and data transformations should be considered for analysis and predictive modeling.

**KRUSKAL WALLIS TEST:**

```python
from scipy.stats import kruskal
import pandas as pd

fetalh = ['baseline value', 'accelerations', 'fetal_movement', 'uterine_contractions',
          'light_decelerations', 'severe_decelerations', 'prolongued_decelerations',
          'abnormal_short_term_variability', 'mean_value_of_short_term_variability',
          'percentage_of_time_with_abnormal_long_term_variability', 'mean_value_of_long_term_variability',
          'histogram_width', 'histogram_min', 'histogram_max', 'histogram_number_of_peaks',
          'histogram_number_of_zeroes', 'histogram_mode', 'histogram_mean', 'histogram_median',
          'histogram_variance', 'histogram_tendency']

# Results will be printed directly
for column in fetalh:
    # Check if the column exists and is numeric
    if column in df.columns and pd.api.types.is_numeric_dtype(df[column]):
        # Prepare the data for the test
        groups = [group[column].dropna() for name, group in df.groupby('fetal_health') if group[column].dropna().size > 0]
        # Check if there is enough variation in the data for each group
        if all(len(set(group)) == 1 for group in groups):
            print(f"Not enough variation in data to perform Kruskal-Wallis test on {column}")
        else:
            # Perform the Kruskal-Wallis test if there are at least two groups with variation
            stat, p = kruskal(*groups)
            print(f"Kruskal-Wallis test for {column}: Statistic = {stat:.2f}, p-Value = {p:.3f}")
    else:
        print(f"{column} is not found in the dataset or is not numeric")
```
